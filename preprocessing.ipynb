```python
import pandas as pd
import numpy as np
import matplotlib_inline as plt
import glob
import os
```

```python
path_Marambaia = glob.glob(r"C:\Users\F8091795\OneDrive - TIM\Projeto-AM\Estacoes-Marambaia/*.csv")
path_Jacarepagua = glob.glob(r"C:\Users\F8091795\OneDrive - TIM\Projeto-AM\Estacoes-Jacarepagua/*.csv")
path_Vila_Militar = glob.glob(r"C:\Users\F8091795\OneDrive - TIM\Projeto-AM\Estacoes-Vila-Militar/*.csv")
df_Marambaia_list = []
df_Jacarepagua_list = []
df_Vila_Militar_list = []
column_mapping =  {
    "HORA (UTC)": "Hora UTC",
    "DATA (YYYY-MM-DD)": "Data",
    "RADIACAO GLOBAL (KJ/m²)": "RADIACAO GLOBAL (Kj/m²)"
    }
for path_arquivo in path_Marambaia:
    df = pd.read_csv(path_arquivo, encoding='latin1', skiprows=8, sep=';')
    df.rename(columns=lambda col: column_mapping.get(col.strip(), col.strip()), inplace=True)
    df_Marambaia_list.append(df)

for path_arquivo in path_Jacarepagua:
    df = pd.read_csv(path_arquivo, encoding='latin1', skiprows=8, sep=';')
    df.rename(columns=lambda col: column_mapping.get(col.strip(), col.strip()), inplace=True)
    df_Jacarepagua_list.append(df)

for path_arquivo in path_Vila_Militar:
    df = pd.read_csv(path_arquivo, encoding='latin1', skiprows=8, sep=';')
    df.rename(columns=lambda col: column_mapping.get(col.strip(), col.strip()), inplace=True)
    df_Vila_Militar_list.append(df)

df_total_Marambaia = pd.concat(df_Marambaia_list, ignore_index=True)
df_total_Jacarepagua = pd.concat(df_Jacarepagua_list, ignore_index=True)
df_total_Vila_Militar = pd.concat(df_Vila_Militar_list, ignore_index=True)
df_total_Marambaia
```

```python
df_total_Jacarepagua = df_total_Jacarepagua.drop(df_total_Jacarepagua.columns[-1], axis=1)
df_total_Marambaia = df_total_Marambaia.drop(df_total_Marambaia.columns[-1], axis=1)
df_total_Vila_Militar = df_total_Vila_Militar.drop(df_total_Vila_Militar.columns[-1], axis=1)
df_total_Jacarepagua
```

```python
from scipy.interpolate import PchipInterpolator

df_mar_copia = df_total_Marambaia.copy()

tempo = df_mar_copia['datetime']

colunas_para_interpolar = [
    "PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)",
    "PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)",
    "PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)",
    "TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)",
    "TEMPERATURA DO PONTO DE ORVALHO (°C)",
    "TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)",
    "TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)",
    "TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)",
    "TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)",
    "UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)",
    "UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)",
    "UMIDADE RELATIVA DO AR, HORARIA (%)",
    "VENTO, DIREÇÃO HORARIA (gr) (° (gr))",
    "VENTO, RAJADA MAXIMA (m/s)",
    "VENTO, VELOCIDADE HORARIA (m/s)",
]

for col in colunas_para_interpolar:
    mask_valid = df_mar_copia[col].notna()
    interpolador = PchipInterpolator(tempo[mask_valid], df_mar_copia[col][mask_valid])
    
    interpolados = interpolador(tempo)
    
    interpolados = np.clip(interpolados, a_min=0, a_max=None)
    
    df_mar_copia.loc[df_mar_copia[col].isna(), col] = interpolados[df_mar_copia[col].isna()]

print(df_mar_copia)
```
